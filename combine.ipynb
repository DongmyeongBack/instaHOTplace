```python
import pandas as pd
df1=pd.read_csv("output1.csv")

lst1=df1.values.tolist()
cnt = 0
for row in lst1:
    if len(row) == 6:
        cnt += 1
    print(row)
print(cnt)
df1
```

```python
df2=pd.read_csv("output2.csv")
lst2=df2.values.tolist()
cnt = 0
for row in lst2:
    if len(row) == 6:
        cnt += 1
    print(row)
print(cnt)
df2
```

```python
def remove_duplicate_names(lst):
    seen_row = set()
    new_lst = []
    for row in lst:
        if row[0] not in seen_row:
            seen_row.add(row[0])
            new_lst.append(row)
    return new_lst
new_lst1 = remove_duplicate_names(lst1)
new_lst2 = remove_duplicate_names(lst2)

new_lst2 = [elem for elem in new_lst2 if elem[0] not in [i[0] for i in new_lst1]]
```

```python
import csv
combined_list = []
index = ['name', 'count', 'address', 'latitude', 'longitude', 'type']
combined_list = [index] + new_lst1 + new_lst2
restaurant = 0
cafe = 0
for i in combined_list:
    if i[5] == 0:
        cafe += 1
    if i[5] == 1:
        restaurant += 1
print(f"restaurant: {restaurant}")
print(f"      cafe: {cafe}")
print(f"     total: {len(combined_list)}")
with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(combined_list)
```
