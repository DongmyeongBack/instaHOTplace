```python
import pandas as pd
df1=pd.read_csv("output1.csv")

lst1=df1.values.tolist()
cnt = 0
for row in lst1:
    if len(row) == 6:
        cnt += 1
    print(row)
print(cnt)
df1
```

```python
df2=pd.read_csv("output2.csv")
lst2=df2.values.tolist()
cnt = 0
for row in lst2:
    if len(row) == 6:
        cnt += 1
    print(row)
print(cnt)
df2
```

```python
for i in lst1:
    for j in lst2:
        if i[0] == j[0]:
            print(i[0])
            print(lst2.index(j))
            ex = lst2.index(j)
```

```python
import csv
combined_list = []
index = ['name', 'count', 'address', 'latitude', 'longitude', 'type']
lst2 = lst2[:ex] + lst2[ex+1:]
combined_list = [index] + lst1 + lst2
for i in combined_list:
    print(i)
with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(combined_list)
```
